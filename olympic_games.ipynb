**Yaş, boy, kilo ve madalya sütunlarında eksik veri olduğunu gördüm. Eksik veri bulunan satır sayısı fazla olduğu için bu satırları silmek istemedim. Bu sebeple yaş, boy ve kilo değişkenindeki boşluklları, ortalama değerler ile doldurdum.**

**Madalya kısmında görünen null değerler aslında eksik veri anlamında değil, o yarışmacının madalya kazanmadığı anlamına geliyor bu sebeple onları da "No_Medal" ifadesi ile doldurdum.**

```python
df.Age.fillna(df.Age.mean(), inplace=True)
```

```python
df.Height.fillna(df.Height.mean(), inplace=True)
```

```python
df.Weight.fillna(df.Weight.mean(), inplace=True)
```

```python
df.Medal.fillna("No_Medal", inplace=True)
```

**Kadın katılımcı sayısının erkeklere oranla çok daha az olduğunu gördüm.**

* **Bu sebeple kadın yarışmacı sayısının yıllar içerisindeki değişimine bakmak istedim. Fakat düzgün bir grafik çıkmadı. Düzeltmek için veri sayısını azaltmayı denedim ama yine de olmadı.**

* **Sonrasında "Year" değişkeni dışındaki başka değişkenler bazında kontrol etmeyi denedim. Bazılarında mantıklı grafikler çıkarken bazılarında çıkmadı.**

```python
Womens = df[(df.Sex == "F") & (df.Medal == "Gold") & (df.Sport == "Athletics") & (df.Season == "Summer")].tail(10)
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Age', data = Womens)
plt.title('Women medals per edition of the Games')
```

* **Bunlar arasındaki farkın sayısal değişkenlik olma durumu olduğunu farkettim. Bu sebeple "Year" değişkeninin türünü datetime'a çevirme işlemini kaldırdım.**

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data = Womens)
plt.title('Women medals per edition of the Games')
```
